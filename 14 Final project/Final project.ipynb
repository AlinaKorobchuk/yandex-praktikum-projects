# Задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Цель проекта:** построить модель, прогнозирующую отток клиентов.

**Входные данные:** информация о договорах клиентов, пресональных данных, об интернет-услугах и об услугах телефонии для каждого клиента.


### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

### Преобразование типов данных признаков и создание новых признаков

<div class="alert alert-block alert-info">

**Комментарий студента:** Преобразуем типы данных некоторых признаков.

```python
#Преобразуем поле begin_date к типу datetime
data['begin_date'] = pd.to_datetime(data['begin_date'], yearfirst=True)
```

<div class="alert alert-block alert-info">

**Комментарий студента:** На основе признака с датой можем создать новый признак - количество дней пользования клиента услугами.

```python
def day_upload(data):
    if data == 'No':
        return '2020-02-01'
    return data
```

<div class="alert alert-block alert-info">

**Комментарий студента:** Добавлю еще дополнительный признак для повышения качества предсказания - год прихода клиента

```python
data['date_upload'] = data['end_date'].apply(day_upload)
data['date_upload'] = pd.to_datetime(data['date_upload'], yearfirst=True)
```

```python
data['quantity_days'] = (data['date_upload'] - data['begin_date']).dt.days
```

```python
data['year'] = data['begin_date'].dt.year
```

```python
#Преобразуем целевой признак: заменим дату на 1, а No на 0 и приведем признак к типу данных int
def change(data):
    if data == 'No':
        return 0
    return 1
```

```python
data['end_date'] = data['end_date'].apply(change)
data = data.rename(columns={'end_date': 'churn'})
```

```python
data.head()
```

```python
data = data.drop(['begin_date', 'date_upload'], axis=1)
```

### Разделение на обучающую и тестовую выборки

```python
#Выделим признаки и целевой признак
features = data.drop(['churn', 'customer_i_d'], axis=1)
target = data['churn']
```

### Кодирование категориальных признаков

<div class="alert alert-block alert-info"> 
    
**Комментарий студента:** Поскольку будем обучать модель градиентного бустинга, которая принимает на вход категориальные признаки, оставим датасет без изменения, но для других моделей создадим другой датасет и закодируем категориальные признаки техникой One-hot encoding (применим параметр drop_first=True, чтобы не попасть в дамми-ловушку).

```python
x_train_category = x_train.copy()
x_test_category = x_test.copy()
```

```python
#Выделим категориальные признаки и поменяем в них тип данных
category_features = ['type', 'paperless_billing', 'payment_method', 'gender', 'partner', 'dependents', 'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_t_v', 'streaming_movies', 'multiple_lines']
for category_feature in category_features:
    x_train_category[category_feature] = x_train_category[category_feature].astype("category")
for category_feature in category_features:
    x_test_category[category_feature] = x_test_category[category_feature].astype("category")
```

```python
encoder = OneHotEncoder(drop='first')
```

```python
#Закодируем x_train
X = encoder.fit_transform(x_train[category_features].values.reshape(-1,14)).toarray()
```

```python
dfonehot = pd.DataFrame(X)
x_train_ohe = pd.concat([x_train.drop(category_features, axis=1).reset_index(drop=True), dfonehot], axis=1)
```

```python
#Закодируем x_test
X = encoder.transform(x_test[category_features].values.reshape(-1,14)).toarray()
```

### Подбор параметров для моделей без учета дисбаланса

```python
#Подберем параметры модели по GridSearchCV, создадим функцию для подбора парметров для 3-х моделей
def selection(model, parameters, x_train, y_train, categorical_feature=None):
    clf = RandomizedSearchCV(model, parameters, verbose=1, scoring='roc_auc', cv=5)
    if model == lgb.LGBMClassifier():
        clf.fit(x_train, y_train, categorical_feature=categorical_feature)
    else:
        clf.fit(x_train, y_train)
    return(clf.best_params_, round(clf.best_score_, 4))
warnings.filterwarnings('ignore') 
```

```python
#Функция для отрисовки roc-кривой
def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic Curve')
    plt.legend()
    plt.show()
```
