# Предсказание заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.
Значение метрики RMSE на тестовой выборке должно быть не больше 48.

<div class="alert alert-block alert-info">
    
**Цель проекта:** построить модель для предсказания количества заказов такси на следующий час.
    
**Входные данные:** исторические данные о заказах такси в аэропортах.    

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from random import randint
random_state = 12345
```

## Подготовка данных

```python
data = pd.read_csv("taxi.csv", parse_dates=[0], index_col=[0])
```

```python
data.head()
```

```python
data.info()
```

<div class="alert alert-block alert-info">
Видим, что всего 26496 записей в исходном наборе данных.

```python
data = data.resample('1H').mean()
```

```python
len(data)
```

<div class="alert alert-block alert-info">
После ресемплирования по одному часу получили временной ряд с 4416 записями.

## Анализ данных 

```python
#Проверим, в хронологическом ли порядке расположен временной ряд
data.index.is_monotonic
```

```python
#Проверим данные на пропуски
data.isna().sum()
```

```python
#Посмотрим, как выглядит временной ряд
data.plot(figsize=(13,8), title="Временной ряд с данными о заказах такси")
plt.show()
```

<div class="alert alert-block alert-info">
Судя по данному графику, можно сказать, что в данных есть аномальные значения. Построим на диаграмму размаха.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=data['num_orders'], orient='h')
plt.xlabel("Количество заказов")
plt.show()
```

<div class="alert alert-block alert-info">
Как правило, то, что находится за пределами усов диаграммы размаха является аномалиями, однако в нашем случае таких значений очень много, есть несколько сильно выбивающихся (больше 50). Эти несколько сильно выбивающихся значений можно заменить средним за последние 3 часа, например.

```python
#Количество значений больше 50
anomalies = data[data['num_orders'] > 50]
len(anomalies)
```

```python
#Посмотрим на аномалии до преобразования
data.loc[anomalies.index, :]
```

```python
dif1 = datetime.timedelta(hours = 1)
dif2 = datetime.timedelta(hours = 2)
dif3 = datetime.timedelta(hours = 3)
for date in anomalies.index:
    data.loc[date, :] = (data.loc[date-dif1, :] + data.loc[date-dif2, :] + data.loc[date-dif3, :]) / 3
```

```python
#Посмотрим на аномалии после преобразования
data.loc[anomalies.index, :]
```

```python
#Посмотрим, как выглядит временной ряд после замены аномалий
data.plot(figsize=(13,8), title="Временной ряд с данными о заказах такси")
plt.show()
```

```python
data.isna().sum()
```

```python
#Произведем декомпозицию временного ряда для того, чтобы посмотреть из каких компонет состоит временной ряд
decomposed = seasonal_decompose(data)
```

```python
plt.figure(figsize=(13, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
```

<div class="alert alert-block alert-info">
Судя по первому графику в данном временном ряде присутствует плавный увеличивающийся тренд, а также явная сезонность, у временного ряда меняется среднее значение. Стационарные временные ряды — это ряды, у которых среднее и стандартное отклонение не меняются. Посчитаем Критерий Дики-Фуллера для того, что определить стационарен или нет временной ряд.

```python
#Посчитаем критерий Дики-фуллера
print ("Критерий Дики-Фулера: p=%f" % adfuller(data)[1])
```

<div class="alert alert-block alert-info">
Критерий Дики-Фуллера меньше 5 %, это говорит о том, что ряд стационарен. Однако, если видим по графику, что ряд не стационарный, критерию не стоит доверять. Произведем дифференцирование ряда(то есть найдем разности), чтобы добиться стационарности ряда, а также добавим новые признаки.

```python
#Создадим функцию для создания новых признаков из даты, а также признаков с разностями и признака скользящего среднего
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data
```

```python
data = make_features(data, 4, 10)
data
```

```python
#Отобразим на графике скользящее среднее
data[['num_orders', 'rolling_mean']].plot(figsize=(13,8))
plt.show()
```

## Обучение моделей

```python
data = data.dropna()
```

```python
features = data.drop('num_orders', axis=1)
target = data['num_orders']
```

```python
#Разделим данные на обучающую, валидационную и тестовую выборки. Принято делать на обучающую, валидационную и тестовую выборки в соотношении 3:1:1. Но так как в задании указано для тестовой выделить 10%, на обучающую оставим - 80 %, на валидационную - 10%.
x_train, x_valid, y_train, y_valid = train_test_split(features, target, shuffle=False, test_size=0.25)
x_valid, x_test, y_valid, y_test = train_test_split(x_valid, y_valid, shuffle=False, test_size=0.5)
```

<div class="alert alert-block alert-info">
Обучим несколько моделей: модель случайного леса, линейной регрессии и градиентного бустинга. Для случайного леса и градиентного бустинга подберем параметры по GridSearchCV.

```python
#Модель случайного леса
model = RandomForestRegressor()
parameters = {
    'random_state': [random_state],
    'n_estimators': [randint(1,50)],
    'max_depth': [randint(1,10)]
}
grid = GridSearchCV(model, parameters, verbose=1)
grid.fit(x_train, y_train)
print(grid.best_params_)
```

```python
model_forest = RandomForestRegressor(random_state=random_state, n_estimators=4, max_depth=2)
model_forest.fit(x_train, y_train)
predictions = model_forest.predict(x_valid)
print("RMSE:", mean_squared_error(y_valid, predictions) ** 0.5)
```

```python
#Модель линейной регрессии
model_regression = LinearRegression()
model_regression.fit(x_train, y_train)
predictions = model_regression.predict(x_valid)
print("RMSE:", mean_squared_error(y_valid, predictions) ** 0.5)
```

```python
#Модель градиентого бустинга
model_boost = lgb.LGBMRegressor()
parameters = {
    'objective': ['regression_l1', 'regression_l2'],
    'random_state': [random_state],
    'n_estimators': [randint(1,100)],
    'learning_rate': [0.03, 0.1],
    'max_depth': [randint(1,20)]
}
grid = GridSearchCV(model_boost, parameters, verbose=1, scoring='neg_mean_squared_error')
grid.fit(x_train, y_train)
print(grid.best_params_)
```

```python
model_boost = lgb.LGBMRegressor(objective='regression_l2', n_estimators=69, max_depth=3, random_state=random_state, learning_rate=0.1)
model_boost.fit(x_train, y_train)
predictions = model_boost.predict(x_valid)
print("RMSE модели градиентного бустинга:", mean_squared_error(y_valid, predictions) ** 0.5)
```

<div class="alert alert-block alert-info">
Наименьшее значение метрики RMSE у модели градиентного бустинга. Поэтому ее выберем для предсказания на тестовой выборке.

## Проверка модели на тестовой выборке

```python
predictions = model_boost.predict(x_test)
print(mean_squared_error(y_test, predictions) ** 0.5)
```

Таким образом, качество модели градиентного бустинга с  подобранными параметрами на тестовой выборке по метрике RMSE равно 9.27.
