# Определение стоимости автомобилей

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. 

**Данные:** В распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей.  
**Цель проекта:** построение модели для определения стоимости. 

Заказчику важны:

- качество предсказания;
- скорость предсказания;
- время обучения.

<div class="alert alert-block alert-info">
Объекты с мощностью выше 2000 л.с. можно удалить, их не так много, а вот объектов с нулевой мощностью довольно много. Эти нули можно заменить медианой по маркам автомобилей, например.

```python
data = data[data['power'] <= 2000]
```

```python
#Заменим нулевые значения в power медианой по группе марок автомобилей
for brand in data['brand'].unique():
    data.loc[(data['brand'] == brand) & (data['power'] == 0), 'power'] = data[(data['brand'] == brand) & data['power'] != 0]['power'].median()
```

```python
columns_isna = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'not_repaired']
for column in columns_isna:
    print(f"Процент пропусков в поле {column}: {round(len(data[data[column].isna()]) / len(data) * 100, 2)} %")
```

<div class="alert alert-block alert-info">
Пропуски в поле vehicle_type, gearbox, fuel_type заполним модой по соответствующим бренду, модели и году регистрации.
Поле model, not_repaired заполним модой по brand и registration_year

```python
#Создадим функцию для создания словарей 
def dicts(feature):
    dct = dict(
    data[data[feature].notna()]
    .groupby(['brand', 'registration_year'])[feature]
    .agg(lambda x: pd.Series(x).mode()[0])
    )
    return dct    
```

```python
model = dicts("model")
not_repaired = dicts("not_repaired")
```

```python
#создадим функцию для заполнения пустых значений в признаках model и not_repaired
def fill(dct, feature):
    for key, value in dct.items():
        data.loc[(data['brand'] == key[0]) & (data['registration_year'] == key[1]) & (data[feature].isnull()), feature] = value
    return data
```

```python
data = fill(model, "model")
data = fill(not_repaired, "not_repaired")
```

```python
#Проверим
data.isna().sum()
```

<div class="alert alert-block alert-info">
Видно, что количество пропусков снизилось значительно в полях model и not_repaired

```python
#Создадим словари для заполнения пропусков в полях vehicle_type, gearbox, fuel_type
def dicts(feature):
    dct = dict(
    data[data[feature].notna()]
    .groupby(['brand', 'model', 'registration_year'])[feature]
    .agg(lambda x: pd.Series(x).mode()[0])
    )
    return dct    
```

```python
vehicle = dicts("vehicle_type")
gearbox = dicts("gearbox")
fuel_type = dicts("fuel_type")
```

```python
#функция для заполнения пропусков по данным полям
def fill(dct, feature):
    for key, value in dct.items():
        data.loc[(data['brand'] == key[0]) & (data['model'] == key[1]) & (data['registration_year'] == key[2]) & (data[feature].isnull()), feature] = value
    return data
```

```python
data = fill(vehicle, "vehicle_type")
data = fill(gearbox, "gearbox")
data = fill(fuel_type, "fuel_type")
```

```python
#Проверим
data.isna().sum()
```

<div class="alert alert-block alert-info">
После заполнения пропусков остались незаполненные объекты. Заполним оставшиеся пропуски модой по бренду и модели

```python
def dicts(feature):
    dct = dict(
    data[data[feature].notna()]
    .groupby(['brand', 'model'])[feature]
    .agg(lambda x: pd.Series(x).mode()[0])
    )
    return dct  
```

```python
vehicle = dicts("vehicle_type")
gearbox = dicts("gearbox")
fuel_type = dicts("fuel_type")
```

```python
def fill(dct, feature):
    for key, value in dct.items():
        data.loc[(data['brand'] == key[0]) & (data['model'] == key[1]) & (data[feature].isnull()), feature] = value
    return data
```

```python
data = fill(vehicle, "vehicle_type")
data = fill(gearbox, "gearbox")
data = fill(fuel_type, "fuel_type")
```

```python
#Посмотрим на зависимость между количественными признаками
sns.pairplot(data[['price', 'registration_year', 'kilometer', 'power', 'number_of_pictures', 'postal_code']])
```

<div class="alert alert-block alert-info">
    
- Линейная зависимость не выявлена, оставляем все признаки.
    
    
    
- Из признаков, содержащих даты, создадим категориальные признаки.    
    
- Для модели линейной регрессии и случайного леса закодируем категориальные признаки техникой OHE. 
    
- Для моделей бустинга категориальные признаки приведем к типу category.   

```python
#Создадим функцию для создания категориальных признаков из полей с датами, которая создает такие признаки для каждого поля: year, month, day, weekday
def category_date(date_feature):
    month = pd.DatetimeIndex(data[date_feature]).month
    year = pd.DatetimeIndex(data[date_feature]).year
    weekday = data[date_feature].apply(lambda x: pd.Timestamp.weekday(x))
    return month, year, weekday, 
```

```python
data_category = data.copy()
```

```python
data_category['month_crawled'], data_category['year_crawled'], data_category['weekday_crawled'] = category_date('date_crawled')
data_category['month_created'], data_category['year_created'], data_category['weekday_created'] = category_date('date_created')
data_category['month_last_seen'], data_category['year_last_seen'], data_category['weekday_last_seen'] = category_date('last_seen')
```

```python
#Проверим
data_category.head()
```

```python
#Кодируем с помощью OHE
data_ohe = pd.get_dummies(data_category, drop_first=True)
```

```python
#Удалим столбцы с датой
data_ohe = data_ohe.drop(['date_crawled', 'date_created', 'last_seen'], axis=1)
```

```python
#Проверим
data_ohe.head()
```

```python
#Категориальные признаки приведем к типу category
category_features = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'not_repaired']
for category_feature in category_features:
    data_category[category_feature] = data_category[category_feature].astype("category")
```

```python
#Удалим столбцы с датой
data_category = data_category.drop(['date_crawled', 'date_created', 'last_seen'], axis=1)
```

```python
#Проверим
data_category.info()
```

## Обучение моделей

<div class="alert alert-block alert-info">
Сначала обучим модель случайного леса и линейной регрессии.
    
    
Разделим данные на обучающую и тестовую выборку, подберем параметры модели по GridSearchCV и сразу выведем оценку качества rmse модели при встроенной cross_val_score.

```python
features_ohe = data_ohe.drop("price", axis=1)
target_ohe = data_ohe['price']
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(features_ohe, target_ohe, test_size=0.4, random_state=random_state)
x_valid, x_test, y_valid, y_test = train_test_split(x_valid, y_valid, test_size=0.5, random_state=random_state)
```

```python
#Подберем параметры для случайного леса и выведем RMSE по кросс-валидации
model_forest = RandomForestRegressor()
parameters = {
    "max_depth": [randint(1,15)],
    "n_estimators": [randint(1,100)],
    "random_state": [random_state]
}
grid = GridSearchCV(model_forest, parameters, verbose=1, scoring='neg_mean_squared_error')
grid.fit(x_train, y_train)
print(grid.best_params_)
```

```python
#Обучим модель, посчитаем качество на валидационной выборке по RMSE
model_forest = RandomForestRegressor(random_state=random_state, max_depth=11, n_estimators=83)
start = time.time()
model_forest.fit(x_train, y_train)
predictions = model_forest.predict(x_valid)
end = time.time()
print("RMSE модели случайного леса:", mean_squared_error(y_valid, predictions) ** 0.5)
print("Время обучения и работы модели случайного леса: {} секунд".format(end-start))
```

```python
#Оценим качество линейной регрессии на валидационной выборке по RMSE
model_regression = LinearRegression()
start = time.time()
model_regression.fit(x_train, y_train)
predictions = model_regression.predict(x_valid)
end = time.time()
print("RMSE модели линейной регрессии:", mean_squared_error(y_valid, predictions) ** 0.5)
print("Время обучения и работы модели линейной регрессии: {} секунд".format(end-start))
```

<div class="alert alert-block alert-info">
Стандартизируем признаки и еще раз оценим качество линейной регрессии на валидационной выборке

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = pd.DataFrame(scaler.transform((x_train)))
scaler.fit(x_valid)
x_valid_scaled = pd.DataFrame(scaler.transform(x_valid))
```

```python
model_regression_scaled = LinearRegression()
start = time.time()
model_regression_scaled.fit(x_train_scaled, y_train)
predictions = model_regression_scaled.predict(x_valid_scaled)
end = time.time()
print("RMSE модели линейной регрессии:", mean_squared_error(y_valid, predictions) ** 0.5)
print("Время обучения и работы модели линейной регрессии: {} секунд".format(end-start))
```

```python
features_boosting = data_category.drop('price', axis=1)
target_boosting = data_category['price']
```

```python
x_train_boost = features_boosting.loc[x_train.index, :]
y_train_boost = target_boosting[y_train.index]
x_valid_boost = features_boosting.loc[x_valid.index, :]
y_valid_boost = target_boosting[y_valid.index]
x_test_boost = features_boosting.loc[x_test.index, :]
y_test_boost = target_boosting[y_test.index]
```

```python
model_boost = lgb.LGBMRegressor()
parameters = {
    'objective': ['regression_l1', 'regression_l2'],
    'random_state': [random_state],
    'n_estimators': [randint(1,100)],
    'learning_rate': [0.03, 0.1],
    'max_depth': [randint(1,20)]
}
grid = GridSearchCV(model_boost, parameters, verbose=1, scoring='neg_mean_squared_error')
grid.fit(x_train_boost, y_train_boost, categorical_feature=category_features)
print(grid.best_params_)
```

```python
model_boost = lgb.LGBMRegressor(objective='regression_l2', n_estimators=72, max_depth=15, random_state=random_state, learning_rate=0.1)
start = time.time()
model_boost.fit(x_train_boost, y_train_boost, categorical_feature=category_features)
predictions = model_boost.predict(x_valid_boost)
end = time.time()
print("RMSE модели градиентного бустинга:", mean_squared_error(y_valid_boost, predictions) ** 0.5)
print("Время обучения модели градиентного бустинга: {} секунд".format(end-start))
```
